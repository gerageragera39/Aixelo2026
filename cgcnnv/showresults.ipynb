```python
import pandas as pd
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt

# Load the CSV
df = pd.read_csv("results/predictions.csv", header=None, names=["id", "predicted", "true"])

# Compute metrics
rmse = mean_squared_error(df["true"], df["predicted"], squared=False)
mae = mean_absolute_error(df["true"], df["predicted"])
r2 = r2_score(df["true"], df["predicted"])

# Print metrics
print(f"RMSE: {rmse:.4f}")
print(f"MAE: {mae:.4f}")
print(f"RÂ² Score: {r2:.4f}")

# Plot predicted vs. true values
plt.figure(figsize=(6,6))
plt.scatter(df["true"], df["predicted"], alpha=0.7)
plt.plot([df["true"].min(), df["true"].max()],
         [df["true"].min(), df["true"].max()],
         color='red', linestyle='--')
plt.xlabel("True Value")
plt.ylabel("Predicted Value")
plt.title("Predicted vs. True Values")
plt.grid(True)
plt.tight_layout()
plt.show()
```
